```python
import pandas as pd
```

```python
df = pd.read_csv('player_game_data.csv')

df.head()
```

```python
def cal_fantasy_points(df):
    points=0
    points += df['points']*1
    points+= df['assists']*1.5
    points+= (df['rebdefensive']+df['reboffensive'])*1.2
    points += (df['blocks']+df['steals'])*3
    points = points - df['turnovers']
    return points
```

```python
df['fantasy_points'] = df.apply(cal_fantasy_points, axis=1)
```

```python
df = df[df['missed']==0]
total_games=df.groupby(['player_name','season','missed'])['nbagameid'].nunique().reset_index(name='total_games')
total_games.sort_values(by='season',ascending=True)
total_games['total_games'].mean()
```

```python
df.sort_values(by='fantasy_points',ascending=False).head(10)
```

```python
total_fantasy_points_per_Player= df.groupby(['player_name','season'])['fantasy_points'].sum().reset_index().sort_values(by='fantasy_points',ascending=False)
total_fantasy_points_per_Player.head(10)
```

```python
total_fantasy_points_per_Player= df.groupby(['player_name'])['fantasy_points'].sum().reset_index().sort_values(by='fantasy_points',ascending=False)
total_fantasy_points_per_Player.head(10).sort_values(by='fantasy_points',ascending=False)
total_fantasy_points_per_Player[total_fantasy_points_per_Player['player_name']=='Julius Randle']
```

```python
contract_df = pd.read_csv('nba_spotrac_data.csv')
contract_df.head(10)
```

```python
contract_df.rename(columns={'player': 'player_name','year':'season'}, inplace=True)
contract_df.head(5)
```

```python
line_up = ["SG","PF","PG","SF","C"]
temp_df = contract_df[contract_df['pos']==line_up[0]]
temp_df.head(5)
```

```python
df_list=[]
for pos_chosen in line_up:
    temp_df = contract_df[contract_df['pos']==pos_chosen]
    fant_table = pd.merge(total_fantasy_points_per_Player,temp_df,on='player_name')
    merged_df = pd.merge(fant_table, total_games[['player_name', 'season', 'total_games']], 
                         on=['player_name', 'season'], how='left')
    merged_df = merged_df[merged_df['total_games']>48.0]
    merged_df['fantasy_points']=(merged_df['fantasy_points']/merged_df['total_games']).round(2)
    df_list.append(merged_df)
```

```python
df_list[1].sort_values(by='fantasy_points',ascending = False)
```
